### 1.2 - Data Preparation ##

Prepare the data for regression task. **(15 Points)**

**Tasks:**
1. Load data for logistic regression.
2. **Generate the SCATTER PLOT of the data**.

**Hints:**
1. The data file is "data_logistic.csv", which are exam scores for students.
2. The data is organized by column: x1 (exam 1 score), x2 (exam 2 score), and label y (pass 1 or fail 0).
3. Please use different colors for postive(label=1) and negative(label=0) data.
4. An example of scatter plots is shown below.
image.png

```python
def plotData(exam1, exam2, color, label):
    plt.scatter(exam1, exam2, color = color, label = label)  

def filterThoseWhoFailedCourse(exam1, exam2, passOrFail):
    failedExam1 = np.array([])
    failedExam2 = np.array([])
    for x in range(len(exam1)):
        if passOrFail[x] == 0:
            failedExam1 = np.append(failedExam1, exam1[x])
            failedExam2 = np.append(failedExam2, exam2[x])
    return failedExam1, failedExam2
            
def filterThoseWhoPassedCourse(exam1, exam2, passOrFail):
    passedExam1 = np.array([])
    passedExam2 = np.array([])
    
    for x in range(len(exam1)):
        if passOrFail[x] == 1:
            passedExam1 = np.append(passedExam1, exam1[x])
            passedExam2 = np.append(passedExam2, exam2[x])
    return passedExam1, passedExam2
    
    
# Data before any filtering
data = pd.read_csv('data_logistic.csv')
exam1 = data.iloc[:, 0]
exam2 = data.iloc[:, 1]
passOrFail = data.iloc[:, 2]

# Data after filtering students who failed the course from those who passed
failedExam1, failedExam2 = filterThoseWhoFailedCourse(exam1, exam2, passOrFail)
passedExam1, passedExam2 = filterThoseWhoPassedCourse(exam1, exam2, passOrFail)

# Students who failed course(red circles), passed course(blue cirlces)
plotData(failedExam1, failedExam2, 'r', 'Failed Course')
plotData(passedExam1, passedExam2, 'b', 'Passed Course')
plt.legend()
plt.show()
```
